```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_boston
```

```python
boston=load_boston()
```

```python
# boston.target.shape
```

```python
# boston.data.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(boston.data,boston.target)
```

```python
x_train,x_valid,y_train,y_valid=train_test_split(x_train,y_train)
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_valid=scaler.transform(x_valid)
x_test=scaler.transform(x_test)
```

```python
input_layer=tf.keras.layers.Input(shape=[13,])
hidden1=tf.keras.layers.Dense(300,activation='relu')(input_layer)
hidden2=tf.keras.layers.Dense(100,activation='relu')(hidden1)
output_layer=tf.keras.layers.Dense(1)(hidden2)

model=tf.keras.models.Model(inputs=[input_layer],outputs=[output_layer])
```

```python
model.summary()
```

```python
model.compile(loss='mse',optimizer='sgd')
```

```python
history=model.fit(x_train,y_train,validation_data=[x_valid,y_valid],epochs=100)
```

```python
y_predict=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_predict)
```
